```python
import os

def find_log_files(directory):
    """
    Find all log files in a given directory and its subdirectories.

    Parameters:
    directory (str): Path to the directory to search.

    Returns:
    list: List of paths to log files.
    """
    log_files = []
    for root, _, files in os.walk(directory):
        for file in files:
            if file.endswith('.log'):
                log_files.append(os.path.join(root, file))
    return log_files

directory_path = '/content/11785-TCR/final_report'
log_files = find_log_files(directory_path)
log_files
```

```python
def extract_metrics_with_labels(file_list, start_marker="#### ", end_marker=".yaml ####"):
    """
    Extract content between specific markers and all labeled sets of metrics from a list of files.

    Parameters:
    file_list (list): List of file paths to process.
    start_marker (str): The marker indicating the start of the content to extract.
    end_marker (str): The marker indicating the end of the content to extract.

    Returns:
    dict: A dictionary where keys are file paths and values are dictionaries containing:
          - 'marked_content': Extracted content between markers.
          - 'metrics': List of dictionaries for each labeled set of metrics.
    """
    extracted_data = {}
    metric_pattern = re.compile(
        r"Accuracy on test data: ([\d.]+)\s+"
        r"AUROC on test data: ([\d.]+)\s+"
        r"AUPR on test data: ([\d.]+)\s+"
        r"F1 Score on test data: ([\d.]+)"
    )
    labels = ["XGB", "XGB(SMI)", "MLP", "MLP(SMI)", "CNN", "CNN(SMI)"]

    for file_path in file_list:
        try:
            with open(file_path, 'r') as file:
                content = file.read()

                # Extract content between markers
                start_idx = content.find(start_marker)
                end_idx = content.find(end_marker, start_idx)
                marked_content = None
                if start_idx != -1 and end_idx != -1:
                    marked_content = content[start_idx + len(start_marker):end_idx].strip()

                # Extract all labeled sets of metrics
                metrics_list = []
                for i, match in enumerate(metric_pattern.findall(content)):
                    metrics = {
                        'Label': labels[i % len(labels)],  # Cycle through labels
                        'Accuracy': float(match[0]),
                        'AUROC': float(match[1]),
                        'AUPR': float(match[2]),
                        'F1 Score': float(match[3])
                    }
                    metrics_list.append(metrics)

                extracted_data[file_path] = {'marked_content': marked_content, 'metrics': metrics_list}
        except Exception as e:
            extracted_data[file_path] = {'error': str(e)}
    return extracted_data

log_file_labeled_data = extract_metrics_with_labels(log_files)
log_file_labeled_data
```

```python
import pandas as pd

def save_metrics_to_csv(extracted_data, output_csv_path):
    """
    Save extracted metrics to a CSV file.

    Parameters:
    extracted_data (dict): The dictionary containing extracted metrics and marked content.
    output_csv_path (str): The path to save the output CSV file.
    """
    # Flatten the data for CSV saving
    rows = []
    for file_path, data in extracted_data.items():
        if 'metrics' in data:
            for metric in data['metrics']:
                rows.append({
                    'File': file_path,
                    'Model': data.get('marked_content'),
                    'Label': metric.get('Label'),
                    'Accuracy': metric.get('Accuracy'),
                    'AUROC': metric.get('AUROC'),
                    'AUPR': metric.get('AUPR'),
                    'F1 Score': metric.get('F1 Score'),
                })

    # Create DataFrame and save to CSV
    df = pd.DataFrame(rows)
    df.to_csv(output_csv_path, index=False)

# Define output path for the CSV file
output_csv_path = "extracted_metrics.csv"

# Save to CSV
save_metrics_to_csv(log_file_labeled_data, output_csv_path)

output_csv_path
```
